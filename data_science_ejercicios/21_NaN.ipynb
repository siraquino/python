##### 2da Forma (Eliminando  NAN)

```python
df = pd.read_csv('20_lichess_tournament_2022.01.24_RHb69yuT_checkmate-coin.csv', index_col = 'Rank')
df.head()
```

```python
df_sin_nan_columna = df.dropna(axis = 1,how = 'all')
df_sin_nan_columna.head()
```

```python
df_sin_nan_columna.head()
```

```python
df_sin_nan_columna.tail()
```

Se puede usar en axis --> **Columns** o **1**

```python
df.dropna(axis='columns',how = 'all')
```

No podemos (en este caso) omitir **how = 'all'** <br>
porque nos elimina la columna Performance y <br>
alli tenemos datos importantes <br>

```python
df.dropna(axis = 1)
```

```python
df.dropna(axis = 1,how = 'all')
```

```python
df.head()
```

```python
df_def = df_sin_nan_columna.dropna(axis = 0)
df_def
```

Podemos usar **axis = 'index'** <br>
o **axis = 0** <br>
para trabajar con las filas <br>

```python
df_def = df_sin_nan_columna.dropna(axis = 'index')
df_def
```
