### Eliminar toda una columna con valor NaN

```python
df_sin_nan_columna = df.dropna(axis = 1,how = 'all')
df_sin_nan_columna
```

### Eliminar toda una fila valor NaN

```python
df_sin_nan_fila = df_sin_nan_columna.dropna(axis = 0)
df_sin_nan_fila
```
